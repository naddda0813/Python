```python
import nltk
import pickle
from nltk.corpus import stopwords  # english 불용어
import re   # 정규표현식을 사용 
nltk.download("all")        #다운로드 전부, 데이터가 많으므로 추후 해당위치에 삭제할것
```

### text를 불러서 word 분류

```python
content = open('../Data/Independence.txt').read()
content
```

```python
# 불필요한 심벌 없애기
cleaned_content = re.sub(r'[^\.\?\!\w\d\s]','',content) # 문장단위로 끊기
cleaned_content
```

```python
# 대문자를 소문자로 변환
cleaned_content = cleaned_content.lower()
cleaned_content
```

```python
# 문장에서 워드를 토큰으로 쪼개기
word_tokens = nltk.word_tokenize(cleaned_content)
print(word_tokens)
```

```python
# 품사 분리 
token_pos=nltk.pos_tag(word_tokens)
# (단어, 품사 )
print(token_pos)
```

```python
# 명사만 추출하기
NN_words = []
for word,pos in token_pos:
    if 'NN' in pos:
        NN_words.append(word)

print(NN_words)
```

```python
# 단어의 원형형태로 복원
# s가 붙은 복수형 등을 원래 단어 사전의 원형태로 돌려줌 
wlem = nltk.WordNetLemmatizer()
lenmatized_words =[]

for word in NN_words:
    new_word = wlem.lemmatize(word)
    lenmatized_words.append(new_word)

print(lenmatized_words)
```

```python
# 불용어 제거 
stopwords_list = stopwords.words('english')
unique_NN_words = set(lenmatized_words)  # 교집합으로 중복값 제거 
final_NN_words = lenmatized_words
# 중복을 제거 하기 위해 복사해서 remove
for word in unique_NN_words:
    if word in stopwords_list:
        while word in final_NN_words:
            final_NN_words.remove(word)
```

```python
print(final_NN_words)
```

```python
# 원하는 단어 제거 
# 예시 
customized_stopwords = ['be','yesterday','today',"it's","don't"]

unique_NN_words1 = set(final_NN_words)
for word in unique_NN_words1:
    if word in customized_stopwords:
        while word in final_NN_words:
            final_NN_words.remove(word)
```

```python
# 빈도분석 
from collections import Counter
# 튜플 형태로 변환 
c = Counter(final_NN_words)
print(c)
```

```python
# 빈도수 높은 순으로 20개 정렬 
print(c.most_common(20))
```

```python
# 워드 클라우드 
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import wordcloud

noun_text =''
for word in final_NN_words:
    noun_text = noun_text + ' ' + word
# 하나의 문장으로 합치기 
noun_text
```

```python
wordcloud = WordCloud(
    max_font_size=60, # 가장 많은 빈도를 보여주는 단어 크기 
    relative_scaling=0.5, # 줄여지는 단어크기 단위
).generate(noun_text)  # generate 가 noun_text의 string value을 입력 받아 count 함 
# generate 단순한 시각화 함수로 카운트 갯수를 우리에게 반환하지는 않음 

# 그리기 
plt.figure(figsize =(12,10))
plt.imshow(wordcloud, interpolation='bilinear')  # interpolation = 써놓은법
# bilinear = 두줄의 선 
plt.axis("off")  # x축 가리기 
plt.show()
```
