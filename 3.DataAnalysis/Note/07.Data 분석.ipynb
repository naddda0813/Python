```python
# fare 
train.Fare.describe()
# 중앙값 50퍼센트에 14.45를  15로 기준 잡고 lowfare(~15),mediumfare(15 ~) , highfare(30~ )인 데이터 프레임 
```

```python
train['Lowfare']  = (train.Fare < 15)
train['Mediumfare'] = (train.Fare >= 15) & (train.Fare < 30) 
train['Highfare'] = (train.Fare >=30)
train[['Fare','Lowfare','Mediumfare','Highfare']]
```

```python
# fare type 추가 
train.loc[train['Lowfare'] == True,'FareType'] = 'Low' 
train.loc[train['Mediumfare'] == True,'FareType'] = 'Medium' 
train.loc[train['Highfare'] == True,'FareType'] = 'High' 
train[['Fare','FareType']]

#train.loc[train['Embarked'] == 'C','Nationality'] = "Austia"
```

```python
# 직접 함수 만들기 
def check_fare_type(fare):  # 컬럼명과 일치 하지 않아도 됨. parameter 값 
    if fare < 15:
        return "Low"
    elif fare >=15 and fare < 30:
        return "Medium"
    else: 
        return "High"
```

```python
# 함수 적용 
train['FareType'] = train['Fare'].apply(check_fare_type)
train[['Fare','FareType']]
```

```python
# age의 NaN을 age의 평균값으로 대체하기(평균나이는 소수점 1자리)
train.Age.mean()
# mean = 29.7
```

```python
# 열과 행을 지정해야만 열 전체에 값이 안들어감
train.loc[train['Age'].isnull(),"Age"] = round(train.Age.mean(),1)
```
