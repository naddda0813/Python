# 행과 열을 이용한 데이터 선택

```python
exam_data = {'수학':[90,80,70], '영어':[98,89,95],
 '음악':[85,95,100], '체육':[100,90,90], '이름': ['서준','우현','인아'] }
df = pd.DataFrame(exam_data)
df
```

```python
# index이름을 정하기
# 지정 컬럼을 인덱스로 쓸거야!
# 조건을 걸었을때도 인덱스가 같이 보이기 때문에 이렇게 설정해주는것의 장점
df.set_index("이름",inplace=True)
df 
```

```python
print(df.loc['서준',['음악','체육']])
print(df.iloc[0,[2,3]])
print(df.iloc[0, 2:4])
print(df.loc['서준','음악':'국어']) # (= print(df.loc['서준','음악':])) 
print(df.iloc[0, 2:])
```

```python
df['국어'] = 80
# df['국어'] = [10,20,30] # 추가할 데이터가 다른경우
df
```

```python
# 행추가 및 데이터 넣기 
df.loc['철수'] = [10,20,30,40,50]
df
```

```python
# 데이터 복사 
df.loc['지희'] = df.loc['인아']
df
```

```python
# 서준의 체육 점수를 80점으로 변경 
df.loc['서준'][3] = 80
```

```python
df
```

```python
# 서준의 음악과 체육점수를 50점으로 변경 
df.loc['서준'][2:4] = 50
df
```

```python
# 서준의 음악 점수를 100점, 체육 점수를 50점으로 변경
df.iloc[0,2] = 100
df.iloc[0,3] = 50
df
```

```python
#2
df.loc['서준',['음악','체육']] = (100,50)
df
```

```python
# 행과 열 바꾸기 - 1 
df.transpose()
```

# 새로운 데이터 프레임


```python
# list를 이용한 DataFrame 생성 

df = pd.DataFrame(
        [[1,4,7,10,13],
        [2,5,8,11,14],
        [3,6,9,12,15],
        ],
        index = ['r0','r1','r2'],        # row names                    
        columns=['c0','c1','c2','c3','c4']  # column names
)
df
```

```python
# 인덱스를 재설정하기 
new_lidex = ['r0', 'r1', 'r2', 'r3', 'r4']
ndf = df.reindex(new_lidex,fill_value=0) # df의 index 재구성 
ndf     # 새로운 index가 생성되고 데이터가 없어 NaN 이 들어옴 
# index를 만들때 부터 fill_value=0
```

```python
ndf = ndf.reset_index()
ndf
# index가 컬럼으로 들어감 
```

```python
df
```

```python
# 내림차순으로 행 인덱스 정렬 
df.sort_index(ascending = False)
```

```python
# 특정 값을 기준으로 열 정렬
df
```

```python

# c1 기준 내림차순 정렬 
df.sort_values(by='c1',ascending=False)
```
