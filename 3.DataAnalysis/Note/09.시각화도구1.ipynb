# Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_excel("../Data/시도별 전출입 인구수.xlsx")
df
```

```python
df.tail()
```

```python
# NaN을 전(Prev) 데이터로 채움 (근처)
df = df.fillna(method='ffill')
df
```

```python
#정보 보기 
df.info()
```

```python
df[df['전출지별'].isnull()]
```

```python
# 서울에서 다른 지역으로 다른 이동한 데이터만 추출하여 정리
#전출지별이 서울특별시이고 전입지별이 서울특별시가 아닌 곳 
#df['전출지별'] == "서울특별시"
#df['전입지별'] != "서울특별시"
df[(df['전출지별'] == "서울특별시") & (df['전입지별'] != "서울특별시") ]
```

```python
df_seoul = pd.DataFrame(df[(df['전출지별'] == "서울특별시") & (df['전입지별'] != "서울특별시") ])
df_seoul
```

```python
df_seoul.shape
```

```python
# 전출지별 컬럼 삭제 
df_seoul.drop('전출지별',axis=1,inplace = True )
# 원데이터에도 실제 삭제를 원하면 inplace = True 
```

```python
df_seoul
```

```python
# 전입지별 -> 전입지 로 컬럼 이름변경
#df_seoul['전입지별'] = '전입지'
# 이렇게 하면 데이터의 내용이 바뀜
df_seoul.rename(columns={'전입지별':'전입지'},inplace = True)
```

```python
df_seoul
```

```python
# df_seoul의 index를 전입지로 변경
df_seoul.set_index('전입지',inplace=True)
df_seoul
```

```python
# 서울에서 경기도로 이동한 인구 데이터 값만 선택
sr_one = df_seoul.loc['경기도']
sr_one
```

```python
plt.plot(sr_one.index, sr_one.values)
plt.show()
```

```python
plt.plot(sr_one)
# 데이터가 하나라 이렇게만 써도 데이터 나옴 
plt.show()
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# 차트 제목 추가
plt.plot(sr_one)
plt.title("서울 -> 경기도 인구 이동 추이 ")

# 이름 추가
plt.xlabel("기간")
plt.ylabel("이동 인구수")
plt.show()
```

```python

# 그래프 꾸미기
# 그림 사이즈 지정(가로 14인치, 세로 5인치)
plt.figure(figsize=(14,5))

# x축 눈금 라벨 회전
plt.xticks(rotation=45)  # 'vertical'01.Series.ipynb은 수직 방향으로 가능 

# y축 라벨 회전
plt.yticks(rotation='vertical')
plt.title("서울 -> 경기도 인구 이동 추이 ")
plt.xlabel("기간")
plt.ylabel("이동 인구수")
plt.plot(sr_one)

# 범례 표시
plt.legend(labels = ['서울->경기'],loc='best')
plt.show()

```

# 동일한 그림에 여러개의 그래프 추가 

```python
# 서울에서 충청남도, 경상북도, 강원도로 이동한 인구데이터값만 선택
df_3=df_seoul.loc[['충청남도','경상북도','강원도']]
df_3
```

```python
plt.style.use('ggplot')
fig=plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)
# 충남
ax.plot(df_3.columns,
    df_3.loc['충청남도'],
    marker='o',
    markerfacecolor='green',
    color='olive',
    linewidth=1.5,
    label='서울->충남'
)
# 경북
ax.plot(df_3.columns,
    df_3.loc['경상북도'],
    marker='o',
    markerfacecolor='blue',
    color='skyblue',
    linewidth=1.5,
    label='서울->경북'
)
# 강원 
ax.plot(df_3.columns,
    df_3.loc['강원도'],
    marker='o',
    markerfacecolor='pink',
    color='lavender',
    linewidth=1.5,
    label='서울->강원'
)


ax.legend(loc='best')
ax.set_title("서울 -> 충남, 경북,강원인구 이동")

ax.set_xlabel("기간")
ax.set_ylabel("이동 인구수")

ax.set_xticklabels(df_3.columns, rotation=90)

# 축 눈금 크기 
ax.tick_params(axis="x", labelsize=5)
ax.tick_params(axis="y", labelsize=5)
plt.shows()
```

```python
# 서울에서 충청남도, 경상북도, 강원도로 이동한 인구데이터값만 선택
df_4=df_seoul.loc[['충청남도','경상북도','강원도','전라남도']]
df_4
```

```python

# 그래프 꾸미기

# style 서식 지정
#plt.style.use("gglot")

# 그래프 객체 생성(figure에 2개의 서브폴롯을 생성)
fig= plt.figure(figsize=(18,14))
ax1 = fig.add_subplot(2,2,1)    # 2행 2열 1번째
ax2 = fig.add_subplot(2,2,2)    # 2행 2열 2번째
ax3 = fig.add_subplot(2,2,3)    # 2행 2열 3번째
ax4 = fig.add_subplot(2,2,4)    # 2행 2열 4번째

ax1.plot(df_4.loc['충청남도'],marker='o',markerfacecolor='red',
    markersize=10,color='pink',linewidth=2,
    label='서울->충남')
ax2.plot(df_4.loc['경상북도'],marker='o',markerfacecolor='yellow',
    markersize=10,color='olive',linewidth=2,
    label='서울->경북')
ax3.plot(df_4.loc['강원도'],marker='o',markerfacecolor='blue',
    markersize=10,color='skyblue',linewidth=2,
    label='서울->강원')
ax4.plot(df_4.loc['전라남도'],marker='o',markerfacecolor='olive',
    markersize=10,color='limegreen',linewidth=2,
    label='서울->전남')


# 제목
ax1.set_title("서울->충남 인구 이동 ")
ax2.set_title("서울->경북 인구 이동 ")
ax3.set_title("서울->강원 인구 이동 ")
ax4.set_title("서울->전남 인구 이동 ")

# ylim
ax1.set_ylim(5000,80000)
ax2.set_ylim(5000,80000)
ax3.set_ylim(5000,80000)
ax4.set_ylim(5000,80000)

#legend
ax1.legend(loc='best')
ax2.legend(loc='best')
ax3.legend(loc='best')
ax4.legend(loc='best')

# x축 눈금 라벨 회전
ax1.set_xticklabels(df_4.loc['충청남도'],rotation=30,size=5)
ax2.set_xticklabels(df_4.loc['경상북도'],rotation=45,size=5)
ax3.set_xticklabels(df_4.loc['강원도'],rotation=45,size=5)
ax4.set_xticklabels(df_4.loc['전라남도'],rotation=45,size=5)

plt.show()
```

### 2010 ~ 2017년의 기간동안 서울에서 각 시도로 이동한 인구의 합계를 구하여 그래프로 비교

```python
# 서울에서 충청남도, 강원도, 경상북도, 전라남도로 이동한 인구 데이터 값만 선택
col_years = list(map(str,range(2010,2018)))
df_4 = df_seoul.loc[['충청남도', '강원도', '경상북도', '전라남도'],col_years]
df_4
```

```python
df_4['합계'] = df_4.sum(axis=1)
# 열방향으로 합쳐 연도의 합
df_4
```

```python
df_total = df_4[['합계']].sort_values(by='합계',ascending=True)
df_total
```

```python
plt.style.use("ggplot")
df_total.plot(kind='barh', #bar는 세로방향 
            color='cornflowerblue',
            figsize= (10,5)
)     
plt.title('서울 -> 타시도 인구 이동')
plt.ylabal('전입지')
plt.xlabal('이동인구수')
plt.show()
```
