# 교촌치킨

```python
# Chrome Browser와 Chrome Driver Version 확인
chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=chrome_options)
driver.get("https://m.kyochon.com/store/store_list")

# 검색 입력창 
xpath = '//*[@id="SEARCH_SHOP_NAM"]'
inputText = driver.find_element(By.XPATH, xpath)
# send spacebar key
inputText.send_keys(' ')
xpath = '//*[@id="btn_search"]'
searchIcon =driver.find_element(By.XPATH, xpath).click()
html=driver.page_source
print(html)
```

```python
#store_list > li:nth-child(1) > h4
#store_list > li:nth-child(1) > div > p > span:nth-child(1)
#store_list > li:nth-child(2) > h4
#container > div:nth-child(4) > div.searched_list
#store_list > li:nth-child(1) > div > p > span:nth-child(1)
```

```python
# 가게이름 
soup = BeautifulSoup(html,"html.parser")
storename =soup.select("ul#store_list li > h4")
print(len(storename))
address =soup.select(".store_info span:nth-child(1) ")
print(len(address))

# 이름 주소 데이터프레임
df = pd.DataFrame(columns=['storename','address'])
for i, name in enumerate(storename):
    # remove first 3 characters from address
    addr = address[i].text.strip()[3:]
    df.loc[i] = [name.text, addr]
print(df)

# save to csv
df.to_csv('../Data/kyochon.csv', encoding='utf-8', mode='w', index=False)
```

```python
# Chrome Browser와 Chrome Driver Version 확인
chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=chrome_options)
driver.get("http://www.pizzaschool.co/bbs/board.php?bo_table=store]")
# 검색창 
xpath = '//*[@id="branch_name1"]'
search_input = driver.find_element(By.XPATH, xpath)

# send key
search_input.send_keys('')

# 검색창 searchicon 클릭 
xpath = '//*[@id="storeSrch1"]/div/div/button'
searchIcon = driver.find_element(By.XPATH, xpath).click()
# 페이지소스 얻기 
html=driver.page_source
print(html)
```

```python
#search1 > li:nth-child(1) > a > dl > dt
#search1 > li:nth-child(1) > a > dl > dd
```

```python
# 가게이름 
soup = BeautifulSoup(html,"html.parser")
storename =soup.select("#search1 > li > a > dl > dt")
#print(storename)
#print(len(storename))
# 주소 
address =soup.select("#search1 > li > a > dl > dd")
print(address)
#print(len(address))
```

```python
# 이름 주소 데이터프레임
df = pd.DataFrame(columns=['storename','address'])
for i, name in enumerate(storename):
    df.loc[i] = [name.text, address[i].text] #storename으로 가져왔어서 name.text는 storename 
print(df)

#save to csv
df.to_csv('../Data/pizzaschool.csv', encoding='utf-8', mode='w', index=False)
```
