```python
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd
```

```python
# Site Address

url = "https://movie.daum.net/boxoffice/yearly?year=2019"
res = req.urlopen(url)
soup = BeautifulSoup(res,"html.parser")
print(soup)
```

```python
#mainContent > div > div.box_boxoffice > ol > li:nth-child(1) > div > div.thumb_cont > strong > a
```

```python
# 타이틀 가져오기 
movielist = soup.select("#mainContent  .link_txt")
print(movielist)
```

```python
movielist
```

```python
# 저장하기 
count = 0 
titleLists = []     # 영화제목 list
countLists = []     # 순위 list

for list in movielist:
    count += 1
    print(count,list.string)
    countLists.append(count)
    titleLists.append(list.string.strip())
```

```python
# list 확인
print(countLists)
print(titleLists)
```

```python
# countlist만으로 dataframe 생성
df = pd.DataFrame(data=countLists,columns=['순위'])
df
```

```python
# 기존 DataFrame에 칼럼과 Lists추가
df['영화제목'] = pd.Series(titleLists)      # 벡터값은 series
df
```

```python
# csv로 저장  
df.to_csv("../Data/DaumMovie.csv", encoding='utf-8', index=False)   
```
