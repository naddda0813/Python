> 현재 layer(층)은 하나 => 입력층은 아니고 은닉층과 출력층, 출력층 하나만 있으므로 

---
# Deep Learning Model 만들기

```python
from tensorflow import keras
```

```python
# 입력층 
input_layer = keras.layers.InputLayer(input_shape=(3,))
# 은닉층
hidden_layer = keras.layers.Dense(256,activation="tanh")
# 출력층 
output_layer = keras.layers.Dense(1,activation="sigmoid")
```

```python
model = keras.Sequential(
    [input_layer,
    hidden_layer,
    output_layer
    ]
)
```

```python
# 손실함수 
model.compile(optimizer="adam",loss='binary_crossentropy',metrics = 'accuracy')
# optimizer : learning rate를 알맞게 조정해줌
# 모두 경사하강법이나 기울기가 jumping 하는 것이다 
# - adam
# - adagrad
# - adadelta
# - rmsprop
# - sgdf : 경사하강법

# loss   => 손실함수
# binary_crossentropy    => 이진분류 및 확률 예측
# metrics = 'accuracy'  => 모델 측정기준 : 정확도
```

```python
# 학습 시키키
import numpy as np
model.fit(np.array(x),np.array(y), epochs=500)
# verbose = 0 => 진행상황 가리기 
```

```python
# hidden layer의 가중치 값출력
hidden_layer.get_weights()[0]
```

```python
hidden_layer.get_weights()[1]
```

```python
# hidden layer 출력 확인
intermediate_layer_model = keras.Model(inputs=model.input,outputs=model.layers[0].output)
intermediate_output = intermediate_layer_model(np.array(x))
```
