### data 정규화

```python
# feature column(weight,height가 한 컬럼 내의 최대 최소 가 차이가 크므로 )
csv['weight'] = csv['weight'] / csv['weight'].max()
csv['height'] = csv['height'] / csv['height'].max()
# 가장 큰 값일 경우 1, 작을 경우 0
```

```python
csv.head()
```

```python
# label
# one hot encoding
bclass = {
    "thin":[1,0,0],
    "normal":[0,1,0],
    "fat":[0,0,1],
}
y= []
for i, v in enumerate(csv['label']):
    y.append(bclass[v])

y
```

```python
# train
X_train = csv[['height','weight']][0:15000]
# 정답
y_train = y[0:15000]
```

```python
X_test= csv[['height','weight']][15001:20001]
#정답
# label
y_test = y[15001:20001]
```

# Deep Learning 만들기

```python
from tensorflow import keras
```

```python
# 층 추가 
model = keras.Sequential()
# input_shape = 입력층
# hidden layer
model.add(keras.layers.Dense(512,activation = "relu",input_shape=(2,)))
# 출력층 
# 다중분류 => class: normal,thin, fat
model.add(keras.layers.Dense(3,activation = "softmax"))
```

```python
from pickletools import optimize
from sklearn import metrics
# optimizer
# 출발 =>  경사하강법 
# 손실함수가 있어야만 optimizer
model.compile(
    optimizer = "rmsprop",
    loss = "categorical_crossentropy", # 출력층이 원핫인코딩이라  sparse는 안쓴다 
    metrics = ['accuracy'] # 모델 측정 방법 : 정확도
)
```

```python
# 데이터 훈련하기 
import numpy as np

model.fit(
    np.array(X_train),  #문제 =>2차원 배열
    np.array(y_train),  # 정답 
    epochs=50
)
```

---
# 테스트 데이터로 평가하기 

```python
# 정확도 
score = model.evaluate(X_test,np.array(y_test))
# y_test => one hot encoding을 해서 list이므로 2차원 으로 바꿈 
# 데이터 손실도 
print("loss :",score[0])    
# 정확도 
print("accuracy:",score[1])
```

> optimizer를 바꾸어 가며 loss(1순위)가 낮을수록 좋은 방법
