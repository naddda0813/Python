### 샘플 차원 변경하기
- 현재 3차원의 배열을 분석을 위해 2차원 배열로 변경
- 배열 중 page 단의 수는 apple, pineapple,banana로 분류하게 변경하기 

```python
apple = fruits[0:100].reshape(-1, 100*100)
pineapple = fruits[100:200].reshape(-1, 100*100)
banana = fruits[200:300].reshape(-1, 100*100)
# 3차원을 하나의 row로 변경
```

```python
apple.shape
```

### 픽셀값 분석하기

```python
# apple의 sample별 평균구하기
apple.mean(axis=1) # 열방향 
```

> 명암을 표시할 때 0은 밝기, 256은 어두움 색 

```python
# 과일별(sample) 빈도수를 히스토그램으로 변환하기
plt.figure(figsize =(15,5))
plt.hist(np.mean(apple,axis=1),alpha=0.8)
plt.hist(np.mean(pineapple,axis=1),alpha=0.8)
plt.hist(np.mean(banana,axis=1),alpha=0.8)  #alpha = opacity
plt.legend(['apple','banana','pineapple'])
plt.show()
```

> 사과와 파인애플의 이미지 구분을 할 수 있을지 확인하려고 히스토그램 그림

```python
# feature(픽셀별) 평균의 막대 그리기 
# 칼럼별 
fig, axs = plt.subplots(1,3,figsize =(20,5))
axs[0].bar(range(10000),np.mean(apple,axis=0))
axs[1].bar(range(10000),np.mean(pineapple,axis=0))
axs[2].bar(range(10000),np.mean(banana,axis=0))
plt.show()
```

```python
# 평균 이미지로 그림을 표현하기 : 이미지로 그리기 위해 다시 2차원 배열로 변경
apple_mean = np.mean(apple,axis=0).reshape(100,100)  # 한장의 이미지 100 *100
pineapple_mean = np.mean(pineapple,axis=0).reshape(100,100)  # 한장의 이미지 100 *100
banana_mean = np.mean(banana,axis=0).reshape(100,100)  # 한장의 이미지 100 *100

fig, axs = plt.subplots(1,3,figsize =(20,5))
axs[0].imshow(apple_mean,cmap='YlGn')
axs[1].imshow(pineapple_mean ,cmap='Wistia')
axs[2].imshow(banana_mean,cmap='gist_earth')
plt.show()
```

> 100장을 사진을 겹친 것과 같음 

### 평균값과 가까운 사진 고르기 
- 위에서 구한 평균 이미지와 각 샘플들의 픽셀값과 비교하여 가장 가까운 거리의 이미지는 해당 과일일 것이다.
- 사과만 잘 분류하는지 시도

```python
abs_diff = np.abs(fruits - apple_mean) # 양수의 음수의 구분이 필요없으므로 절대값 사용
abs_mean = np.mean(abs_diff,axis=(1,2)) # abs_diff가 3차원이기 때문에 page 차원을 제외하고 행과열 평균을 구한다
abs_mean.shape  # 300개의 원소로 이루어진 1차원 배열
```

> abs_mean과 apple의 차이가 가장 작은 것으로 구분 => apple로 예측

```python
# 300개의 데이터 중 차이가 적은 순으로 정렬하면서 해당 이미지를 불러오기 위한 index 필요(arg.sort)
apple_index = np.argsort(abs_mean)[:100]
apple_index
```

```python
fig, axs = plt.subplots(10,10,figsize =(10,10))
for i in range(10):
    for j in range(10):
        axs[i,j].imshow(fruits[apple_index[i*10+j]],cmap='YlGn')
        axs[i,j].axis("off")    # 축 안보이기 
plt.show()
```

> 이와 같은 방법으로 비지도학습인 군집분류 알고리즘이 형성된다
