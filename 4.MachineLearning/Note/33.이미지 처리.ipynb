# numpy를 image로 변환하기 

```python
import numpy as np
fruits = np.load("../Data/fruits.npy")
# npy는 np.load
fruits.shape
```

```python
# image 변환
from PIL import Image
for i in range(0,300):
    image = Image.fromarray(fruits[i]).convert("L")
    image.save("../Data/fruits/fruit_%3d.jpg"%i,format="JPEG")
```

## image를 읽어서 numpy 변환

```python
import glob
```

```python
numberOfData = 300
imgSize = 100
```

```python
# 배열 초기화 
fruitsData =np.zeros(numberOfData* imgSize*imgSize,dtype=np.int32).reshape(numberOfData,imgSize,imgSize)
```

```python
i = 0 
for file in sorted(glob.glob("../Data/fruits/*.jpg")):
    img = np.array(Image.open(file),dtype=np.int32)
    fruitsData[i,:,:] = img
    i +=1
```

```python
# image 확인
import matplotlib.pyplot as plt
plt.imshow(fruitsData[0].reshape(100,100))
plt.show()
```

```python
np.save("../Data/fruitsData",fruitsData)
```
