```python
test_data = test.loc[:,test.columns != 0]
test_label = test.loc[:,test.columns == 0]
```

```python
# 학습하기
from sklearn import svm
clf = svm.SVC()
clf.fit(train_data,train_label)
```

```python
# 정확도 
clf.score(test_data,test_label)
```
