# 머신러닝이란? 
: 수많은 데이터를 학습시켜 거기에 있는 패턴을 찾아내고 이를 기반으로 데이터를 분류하거나 예측 
- 일반 소프트웨어(결정론적 소프트웨어)는 언제나 똑같은 과정을 거쳐 똑같은 결과를 냅니다.
- 머신러닝(비결정론적 소프트웨어) 기반 소프트웨어는 데이터와 알고리즘에 따라 때로는 다른 결과를 추론해 냅니다.

: 추론 : 이미 알고있는 것으로부터 논리적 결론을 도출하는 과정

### 결정론적 소프트웨어와 비결정론적 소프트웨어의 차이점 
: 예를 들어 지문인식기의 소프트웨어를 결정론적 소프트웨어로 처리하였다고 가정하면 지문입력기에 지문이 조금만 틀어줘도 거부 메세지가 나온다. 
이에 반해 비결정론적 소프트웨어는 지문의 위치나 지문이 조금 달라져도 추론을 통해 지문의 가부를 결정합니다

### 머신러닝 개요
: 머신러닝이란 인공지능 연구 과재 중의 하나로 인간의 뇌가 자연스럽게 수행하는 '학습'이라는 능력을 컴퓨터로 구현하는 방법
- 어느 정도 규모가 있느 샘플 데이터를 입력으로 넣어 분석 시킨다.
- 이때 데이터에서 일정한 규칙을 찾아낸다
- 그리고 찾아낸 규칙을 기반으로 다른 데이터를 뷴류(classification)하거나 예측(regression)하는 것.


### 머신러닝
##### 1)지도학습(supervised learning)
: 학습 시 데이터와 답을 같이 학습한 후 다른 데이터의 값을 추출
##### 2)비지도학습(unsupervised learning)
: 학습 시 데이터만 학습한 후 다른 데이터와 규칙성을 찾기. 본질적인 구조 등을 확인할 때 사용
##### 3)강화학습(reinforcement learning)
: 부분적으로 답을 입력한 후 데이터를 기반으로 최적의 답을 찾기
##### 4)딥러닝(Deep learning)
: 지도학습과 비지도 학습의 관련이 있으나 특이성은 학습자료에 대한 인간의 개입을 받지 않는다.

# 머신러닝의 종류
(1) 데이터 수집 <br>
(2) 데이터 가공 <br>
(3) 데이터 학습 <br>
  (3-1) 학습 방법 선택 <br>
  (3-2) 매개 변수 조정 <br>
  (3-3) 모델학습 <br>
(4) 모델 평가 <br>
(5) 예측력이 좋은 경우 다음 단계 진행, 아니면 (3)번부터 다시 <br>
(6) 성공

# 머신러닝 응용분야
### 클래스 분류 (classification)
- 특정 데이터에 레이블(label)을 붙여 분류할 수 있습니다.
- 스팸 메일 분류, 필기인식, 증권 사기 방지 등 

### 클러스터링 (clustering) : 그룹나누기
- 값의 유사성을 기반으로 데이터를 여러 그룹으로 나누기
- 사용자의 취향을 그룹으로 묶어 사용자 취햡에 맞는 광고 제공

### 추천(Recommendation)
- 특정 데이터를 기반으로 다른 데이터를 추천하는 것
- 인터넷 서점에서 구매한 책들을 기반으로 다른 책 추천 

### 회귀 (Regression)
- 과거의 데이터를 기반으로 미래를 예측한 것입니다
- 판매 예측, 주가변동

### 차원 축소 (Dimensionality Reduction)
- 데이터의 특성을 유지하면서 데잍어의 양을 줄이는 것입니다. 
- 데이터를 시각화하거나 구조를 추출해서 용량을 줄여 계산을 빠르게 하거나 메모리를 절약할 때 사용

# 프로젝트 과정
: 머신러닝 프로젝트는 크게 문제정의 ,데이터 획득, 모델 구현, 검증 ,테스트의 단계로 구현

### 문제 정의 
- 프로젝트의 목표를 정의하고 그에 따른 계획을 세우는 단계
- 대략적으로 어떤 데이터를 수지해서 어떤 머신러닝 알고리즘으로 문제를 해결할지 결정하는 단계

### 데이터 획득
- 머신러닝 모델 학습에 필요한 데이터를 수집하는 단계
- 성공적인 머신러닝 프로젝트는 뛰어난 머신러닝 실려곱다도 얼마나 충분히 데이터를 획득했느냐에 의해 결정이 됩니다..
- 획득한 데이터는 학습데이터, 검증데이터, 테스트 데이터로 구분되어 사용 합니다. 

### 모델 구현
- 학습 데이터를 기반으로 한 한개 이상의 머신러닝 모델을 구현

### 검증
- 검증 데이터를 사용해 구현된 머신러닝 모델들의 성능을 검증
- 최종적으로 검증 결과가 가장 높은 모델 하나를 선택

### 테스트
- 검증 결과를 통해 선택된 모델의 성늘을 테스트 데이터로 측청 (일반화 성능 측정)
